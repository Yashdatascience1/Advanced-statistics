# Case Study 
## The annual Percentage Rate of Interest Charged on New-Car Loans at Nine of the 					Largest Banks in Six American Cities is provided and It is believed that the mean interest rate is same across the six american cities.  Conduct the hypothesis test and provide your inference


Use level of significance $\alpha$ = 0.05

```python
import pandas as pd
import numpy as np
from statsmodels.formula.api import ols
from statsmodels.stats.anova import _get_covariance,anova_lm
import seaborn as sns
from scipy.stats import shapiro
from scipy.stats import levene
import scipy.stats as stats
```

```python
df_melt = pd.melt(df,var_name="City",value_name="Interest_Rate")
#df_melt = pd.melt(df)
```

# 1. Postulate the Null and Alternate Hypothesis

H0 : The mean interest rate of all the 9 banks across the cities are equal


H1 : The mean interest rate of all the 9 banks across the cities aren't equal

# 2. Perform the analysis of variances test on the dataset

```python
formula = "Interest_Rate~C(City)"
model = ols(formula,df_melt).fit()
aov_table = anova_lm(model)
print(aov_table)
```

# 3. What are sum of squared variance within and sum of squared variance between?

```python
Sum_of_Squared_Variance_Between=aov_table.loc["C(City)","sum_sq"]
print("Sum of Squared Variance between the cities is {}".format(Sum_of_Squared_Variance_Between))
```

```python
Sum_of_Squared_Variance_Within = aov_table.loc["Residual","sum_sq"]
print("Sum of Squared Variance within the cities is {}".format(Sum_of_Squared_Variance_Within))
```

# 4. Calculate sum of squared total for above

```python
Sum_of_Squared_total = Sum_of_Squared_Variance_Within + Sum_of_Squared_Variance_Between
print("Sum of Squared total is {}".format(Sum_of_Squared_total))
```

# 5. What is the significance of F dist/statistic?

It gives the degree of how relatively greater the difference is 'between group means' (MSB) compared to 'within group variance' (MSW)

# 6. Interpret the results and present the conclusion

Since pvalue of the F test is less than 0.05, hence we can reject the null hypothesis that the mean interest rate of all the 9 banks across the cities are equal
